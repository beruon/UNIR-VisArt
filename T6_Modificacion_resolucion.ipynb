```python
# Importar Librerías
import cv2
import numpy as np
import matplotlib.pyplot as plt
import requests
from io import BytesIO
from google.colab import files
```

# Redimensionar Imagen

```python
# Función para redimensionar la imagen
def resize_image(image, size):
    return cv2.resize(image, (size, size), interpolation=cv2.INTER_AREA)
```

```python
# Opción 1: Descargar la imagen de Lenna desde una URL
url = "https://upload.wikimedia.org/wikipedia/en/7/7d/Lenna_%28test_image%29.png"
response = requests.get(url)
image = np.array(bytearray(response.content), dtype=np.uint8)
image = cv2.imdecode(image, cv2.IMREAD_GRAYSCALE)  # Cargar en escala de grises

# Opción 2: Cargar la imagen desde un archivo local en Jupyter
# image = cv2.imread('ruta_a_la_imagen/lenna.png', cv2.IMREAD_GRAYSCALE)

# Opción 3: Cargar la imagen desde un archivo local en Google Colab
# from google.colab import files
# uploaded = files.upload()
# image_path = list(uploaded.keys())[0]
# image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
```

```python
# Crear las versiones redimensionadas
sizes = [256, 128, 64, 32]
resized_images = [resize_image(image, size) for size in sizes]

# Crear una cuadrícula 2x2
fig, axs = plt.subplots(2, 2, figsize=(10, 10))

# Añadir las imágenes redimensionadas a la cuadrícula
for i, ax in enumerate(axs.flat):
    ax.imshow(resized_images[i], cmap='gray')  # Usar cmap='gray' para escala de grises
    ax.set_title(f'{sizes[i]}x{sizes[i]}')
    ax.axis('off')

# Mostrar la cuadrícula
plt.tight_layout()
plt.show()
```

# Píxel en Escala de Grises

```python
# Opción 1: Descargar la imagen de Lenna desde una URL
# url = "https://upload.wikimedia.org/wikipedia/en/7/7d/Lenna_%28test_image%29.png"
# response = requests.get(url)
# image = np.array(bytearray(response.content), dtype=np.uint8)
# image = cv2.imdecode(image, cv2.IMREAD_GRAYSCALE)  # Cargar en escala de grises

# Opción 2: Cargar la imagen desde un archivo local en Jupyter
# image = cv2.imread('ruta_a_la_imagen/lenna.png', cv2.IMREAD_GRAYSCALE)

# Opción 3: Cargar la imagen desde un archivo local en Google Colab
uploaded = files.upload()
image_path = list(uploaded.keys())[0]
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

# Redimensionar la imagen a 16x16 píxeles
resized_image = cv2.resize(image, (16, 16), interpolation=cv2.INTER_AREA)

# Crear la figura
plt.figure(figsize=(8, 8))
plt.imshow(resized_image, cmap="gray")
plt.title("Imagen en escala de grises (16x16) con valores de intensidad")
plt.axis("off")

# Añadir los valores de intensidad en cada píxel
for i in range(resized_image.shape[0]):
    for j in range(resized_image.shape[1]):
        value = resized_image[i, j]
        plt.text(j, i, f"{value}", color="red", ha="center", va="center", fontsize=10)

# Mostrar la imagen con los valores
plt.show()
```

# Píxel en Modelo RGB

Por defecto, cv2.imread carga imágenes en el formato BGR (Blue-Green-Red), y necesitas convertirlo explícitamente a RGB para que los colores se vean correctamente en matplotlib.

```python
# Opción 1: Descargar la imagen de Lenna desde una URL
# url = "https://upload.wikimedia.org/wikipedia/en/7/7d/Lenna_%28test_image%29.png"
# response = requests.get(url)
# image = np.array(bytearray(response.content), dtype=np.uint8)
# image = cv2.imdecode(image, cv2.COLOR_BGR2RGB)

# Opción 2: Cargar la imagen desde un archivo local en Jupyter
# image = cv2.imread('ruta_a_la_imagen/lenna.png')

# Opción 3: Cargar la imagen desde un archivo local en Google Colab
uploaded = files.upload()
image_path = list(uploaded.keys())[0]
image = cv2.imread(image_path)

# Convertir la imagen de BGR a RGB
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Redimensionar la imagen a 16x16 píxeles
resized_image = cv2.resize(image, (8, 8), interpolation=cv2.INTER_AREA)

# Convertir la imagen a un array de numpy
image_array = np.array(resized_image)

# Crear la figura
plt.figure(figsize=(8, 8))
plt.imshow(image_array)
plt.title("Imagen a color (8x8) con valores RGB")
plt.axis("off")

# Añadir los valores RGB en cada píxel
for i in range(image_array.shape[0]):
    for j in range(image_array.shape[1]):
        r, g, b = image_array[i, j]
        text = f"{r},{g},{b}"
        # Colocar el texto con un color visible (por ejemplo, blanco con borde negro)
        plt.text(j, i, text, color="red", ha="center", va="center", fontsize=6)

# Mostrar la imagen con los valores RGB
plt.show()
```
